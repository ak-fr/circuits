```python
from math import log2

# Implementation copied from "An efficient multiplier for large numbers"
def karatsuba(A, B, n):
    print(f"Inside: n={n}, A={A:#0b}, B={B:#0b}")
    if n == 1:
        return A & B

    mask = (1 << (n//2)) - 1 # lower half bits
    A_H, B_H = A >> (n//2), B >> (n//2)
    A_L, B_L = (A&mask), (B&mask)

    print(f"A = {A : #0b}, A_L = {A_L : #0b}, A_H = {A_H : #0b}")
    print(f"B = {B : #0b}, A_L = {B_L : #0b}, A_H = {B_H : #0b}")
    print(f"mask = {mask : #0b}")

    
    P1 = karatsuba(A_H, B_H, n//2)
    P2 = karatsuba(A_L, B_L, n//2)
    # P3 is problematic because . There should be a speical treatment.
    # https://github.com/JC-S/Karatsuba_multiplier_HDL/ Splitted (A_H + A_L) into a bit and a 
    P3 = karatsuba(A_H + A_L, B_H + B_L, n//2)

    print(f"n={n}: P1 = {P1}, P2 = {P2}, P3 = {P3}")

    m = n//2
    return 2**(2 * n) * P1 + 2**m * (P3 - P2 - P1) + P2
    
```

```python
A, B, n = 7, 5, 4
karatsuba(A, B, n), bin(A*B), len(bin(A*B)) - 2
```
