```python
from math import log2

# Implementation copied from "An efficient multiplier for large numbers"
def karatsuba(A, B, n):
    print(f"Inside: n={n}, A={A:#0b}, B={B:#0b}")
    if n == 1:
        return A & B

    m = n//2
    mask = (1 << m) - 1 # lower half bits
    A_H, B_H = A >> m, B >> m
    A_L, B_L = (A&mask), (B&mask)

    print(f"A = {A : #0b}, A_L = {A_L : #0b}, A_H = {A_H : #0b}")
    print(f"B = {B : #0b}, A_L = {B_L : #0b}, A_H = {B_H : #0b}")
    print(f"A_L - A_H = {A_L - A_H}, B_H - B_L = {B_H - B_L}")
    print(f"mask = {mask : #0b}")

    
    Z2 = karatsuba(A_H, B_H, n//2)
    Z1 = karatsuba(abs(A_L - A_H), abs(B_H - B_L), n//2)
    s = 1
    if (A_L < A_H):
        s = s*-1
    if (B_H < B_L):
        s = s * -1
    Z0 = karatsuba(A_L, B_L, n//2)
    # P3 is problematic because (A_H + A_L)  have length n/2 + 1.
    # https://github.com/JC-S/Karatsuba_multiplier_HDL/ Splitted (A_H + A_L) into a bit and an n bit number 

    print(f"n={n}, n/2={m}, s={s}: Z2 = {Z2}, Z1 = {Z1}, Z0 = {Z0}, val={ 2**n * Z2 + 2**m * (s*Z1 + Z2 + Z0) + Z0}")

    return 2**n * Z2 + 2**m * (s*Z1 + Z2 + Z0) + Z0
    
```

```python
A, B, n = 35, 99, 8
karatsuba(A, B, n), A*B, len(bin(A*B)) - 2
```
